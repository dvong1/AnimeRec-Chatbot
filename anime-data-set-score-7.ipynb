```python
import numpy as np  # Numerical operations
import pandas as pd  # Data manipulation

# Load the dataset into Pandas
data_path = "/kaggle/input/myanimelist-dataset/anime-dataset-2023.csv"
df = pd.read_csv(data_path)

# Convert the 'Score' column to numeric from str value, coercing errors to NaN
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')

# Drop rows where 'Score' is NaN this includes any errors from previous line
df = df.dropna(subset=['Score'])

# Filter rows where 'Score' > 7
filtered_df = df[df['Score'] > 7]

# Resets the index after the dropping NaN
filtered_df = filtered_df.reset_index(drop=True)

#Saves the filtered subset to a new CSV called filtered_anime
filtered_df.to_csv('/kaggle/working/filtered_anime.csv', index=False)
```
